** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
#get 2013
sal1=sal[sal['Year']==2013] 
df=sal1[['JobTitle','Year']].sort_values('JobTitle')
```

```python
df.head()
```

```python
#Groupby
df_count=df.groupby('JobTitle').count()
df_count.head(10)
```

```python
#count number of unique JobTitle
(df_count.Year==1).sum()
```
