### Object Oriented Method

```python
fig = plt.figure()
```

```python
#figure object imagery blank canvas
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.2,0.8,0.8])
axes2 = fig.add_axes([0.5,0.3,0.4,0.4])
#first two numbers are shift, while the next two size (enlargement/shrinkage)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.2,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('BIGGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
#first two numbers are shift, while the next two size (enlargement/shrinkage)
```

### Iterate through the axes

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #does the space automatically 
#axes.plot(x,y)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#Plot on first axes
axes[0].plot(x,y)
axes[0].set_title('First Plot')
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#Plot on second axes
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#Plot on first axes
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2)) #,dpi=100)
ax = fig.add_axes ([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig,axes = plt.subplots(figsize=(8,2)) #,dpi=100)
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2)) #,dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Legends

```python
fig = plt.figure() #,dpi=100)
ax = fig.add_axes ([0,0,1,1])

ax.plot(x,y)
ax.plot(y,x)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure() #,dpi=100)
ax = fig.add_axes ([0,0,1,1])

ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
fig = plt.figure() #,dpi=100)
ax = fig.add_axes ([0,0,1,1])

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label= 'X Cubed')

ax.legend(loc=10) 
#loc = 0 best
#loc = 10 center
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=4) #lw also works

#RGB Hex Code
#ax.plot(x,y,color='#FF9C00') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=4, alpha=0.5)

#RGB Hex Code
#ax.plot(x,y,color='#FF9C00') 
```

```python
### Linestyle
```

```python
# :, --, step, dot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', linewidth=4, linestyle='--')

#RGB Hex Code
#ax.plot(x,y,color='#FF9C00') 
```

```python
### Marker
```

```python
# :, --, step, dot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', linewidth=1, linestyle='-', marker = 'o', markersize=10)

#+
#o
```

```python
# :, --, step, dot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', linewidth=1, linestyle='-', marker = 'o', markersize=10,
       markerfacecolor= 'yellow', markeredgewidth=3, markeredgecolor='purple')

#+
#o
#s
```

```python
# :, --, step, dot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', linewidth=1, linestyle='-', marker = 'o', markersize=10,
       markerfacecolor= 'yellow', markeredgewidth=3, markeredgecolor='purple')

ax.set_xlim([0,3])
ax.set_ylim([0,10])

#+
#o
#s
```
