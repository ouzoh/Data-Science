# Pandas Built-in Data Visualization

In this lecture we will learn about pandas built-in capabilities for data visualization! It's built-off of matplotlib, but it baked into pandas for easier usage!  

Let's take a look!

More for quick analysis

Seaborn and MatLib is better

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head() #time series index
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head() #non-time series index (random data with sequential index)
```

```python
df1['A'].hist() #it calls matplotlib
```

```python
import seaborn as sns #makes the plot look nicer
```

```python
df1['A'].hist(bins=30) 
```

## Alternative Way
*plot(kind=' ')* **or** 
*plot.hist( )*

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist()
```

```python
df2.head()
```

```python
df2.plot.area(alpha=0.4) #alpha for transparency
```

```python
df2.plot.bar() # bar plot index value as category 
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1.plot.line(x=df1.index,y='B', figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B', c= 'C',cmap ='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B', s= df1['C']*100)
```

```python
df1.plot.box() #(x='A',y='B', s= df1['C']*100)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

```python
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```

```python
df2.plot.density()
```
